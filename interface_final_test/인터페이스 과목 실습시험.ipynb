- 본 프로그램은 섭씨를 화씨로 변환 해주는 프로그램입니다.
- 섭씨는 화씨로 변환하는 공식은 ((9/5) * 섭씨온도) + 32
- 출력되는 온도는 소수점 이하 둘째 자리까지 출력 되도록 합니다. 
- << 프로그램 결과>> 
- 변환하고 싶은 섭씨 온도를 입력해 주세요
- 32.2 
- 섭씨온도 : 32.2
- 화씨온도 : 89.96

```python
print('변환하고 싶은 섭씨 온도를 입력해 주세요')
subssi = float(input())
print('섭씨 온도 : ',subssi)
hwassi =  ((9/5) * subssi) + 32
print('화씨 온도 : {0:.15f}'.format(hwassi))
print('화씨 온도 : ', round(hwassi,2))
```

1. Pandas를 사용하여 data/data_draw_korea.csv File Read 후 데이터 분석하기.
- pandas import 하기
- read_csv() 함수로 csv file read
- 상위 5개의 데이터와 하위 5개의 데이터를 확인한다.
- ‘인구수’ 컬럼값의 최소, 평균, 최대값을 구한다.
- ‘광역시도’ 컬럼값의 unique 한 값을 구한다. 
- ‘광역시도’ 컬럼값 별로 row를 count 한다. 
- 광역시도 중 ‘서울특별시’ 데이터만 출력한다. 
- ( 컬럼은 모두 출력, 정렬은 인구수별로, DESC )

```python
# pandas import 하기
import pandas as pd
# read_csv() 함수로 csv file read
data = pd.read_csv('data/data_draw_korea.csv')
# 상위 5개의 데이터
data.head()
# 하위 5개의 데이터
data.tail()
# ‘인구수’ 컬럼값의 최소, 평균, 최대값을 구한다.
data['인구수'].min()
data['인구수'].mean()
data['인구수'].max()
#‘광역시도’ 컬럼값의 unique 한 값을 구한다.
data['광역시도'].unique()
#‘광역시도’ 컬럼값 별로 row를 count 한다. 
data['광역시도'].value_counts()
# 광역시도 중 ‘서울특별시’ 데이터만 출력한다. ( 컬럼은 모두 출력, 정렬은 인구수별로, DESC )
data.loc[data['광역시도'] == '서울특별시']
```

1. requests와 beautifulsoup을 사용하여 네이버 뉴스 기사 제목 출력하기.
- requests, beautifulsoup, urljoin import 하기
- url = 'https://news.naver.com/'
- http status code 출력한다.
- 뉴스기사 제목과 링크를 출력한다. 
- 뉴스기사를 선택하는 CSS 선택자는 a[href*=read.nhn] 이며 select() 함수를 사용한다.
- 뉴스기사 링크는 상대 패스 이므로 절대 패스 형태로 출력하세요(urljoin() 함수 사용)

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

url = 'https://news.naver.com/'
response = requests.get(url)
print('상태코드 : ',response.status_code)
soup = BeautifulSoup(response.text, 'html.parser')
tag_list = soup.select('a[href*=read.nhn]')
for a_tag in tag_list:
    print('기사 제목 : ', a_tag.text.strip())
    print('기사 링크', urljoin(url, a_tag['href']))
```
