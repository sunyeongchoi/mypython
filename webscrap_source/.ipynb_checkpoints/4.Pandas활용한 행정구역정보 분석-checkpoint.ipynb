### read_csv() 함수
* csv file읽기
* head(), tail() 함수로 일부 데이터 가져오기
* columns, index, shape, info()로 meta 정보 확인
* describe(), max(), min(), mean() 집계함수
* 컬럼명을 변경하고, 특정 컬럼을 인덱스로 변경하기

```python
import pandas as pd

data = pd.read_csv('data/data_draw_korea.csv')
print(type(data))
# head() 함수의 default=5
data.head(3)
```

```python
data.tail()
```

```python
data.tail(10)
```

```python
# 컬럼명
print(data.columns)
print('--------------------')
# 인덱스
print(data.index)
print('--------------------')
# 몇행 몇열
print(data.shape)
```

```python
# info() - DataFrame의 meta정보
data.info()
```

```python
# describe() - 집계함수 (문자열 타입 빼고 숫자타입에 대해서만)
data.describe()
```

```python
# 인구수 컬럼값의 최대, 최소, 평균
print(type(data['인구수']))
print(data['인구수'].head())
print('최소',data['인구수'].min())
print('평균',data['인구수'].mean())
print('최대',data['인구수'].max())
print('표준편차',data['인구수'].std())
```

```python
# 면적 컬럼값의 최대, 최소,평균
print('최소', data['면적'].min())
print('평균', data['면적'].mean())
print('최대', data['면적'].max())
```

```python
# 컬럼명 변경
data = data.rename(columns={'Unnamed: 0':'seq'})
data.head(2)
```

```python
# seq컬럼을 index로 변경하기
data = data.set_index('seq')
data.head(2)
```

```python
# 부산광역시의 행정구역, 인구수 컬럼을 가져오기
busan = data['광역시도']=='부산광역시'
data.loc[busan,['행정구역','인구수']]
```

```python
# 인구수 평균
pop_mean_value = data['인구수'].mean()
pop_mean_value 
```

```python
# 면적에 대한 평균값
area_mean_value = data['면적'].mean()
area_mean_value
```

```python
# 인구수가 평균보다 낮은 행정구역 가져오기
data.loc[data['인구수'] < pop_mean_value]
```

```python
# 인구수가 평균보다 높은 행정구역 가져오기
data.loc[data['인구수'] > pop_mean_value]
```

```python
# 인구수가 평균보다 높고, 면적도 평균보다 높은 지역 구하기
data.loc[(data['인구수'] > pop_mean_value) & (data['면적'] > area_mean_value)]  
```

```python
# 경기도에서 인구수가 평균보다 높은 행정구역 구하기
data.loc[(data['광역시도'] == '경기도') & (data['인구수'] > pop_mean_value),'행정구역']
```

```python
# 강원도에서 가장 높은 인구수는?
data.loc[(data['광역시도'] == '강원도'),'인구수'].max()
```

```python
# 람다함수
people_count = lambda area: data.loc[(data['광역시도'] == area),'인구수'].max()
people_count('강원도')
```

```python
# 힘수
def people_count(area):
    return data.loc[(data['광역시도'] == area),'인구수'].max()
```

```python
people_count('강원도')
```

```python
data.loc[data['인구수'] == people_count('강원도')]
```

```python
get_pop_max = lambda area : data.loc[data['인구수'] == people_count(area),['광역시도','행정구역','인구수']]
get_pop_max('부산광역시')
```

```python
type(data['광역시도'].unique()) #numpy.ndarray는 동일한 타입만 넣을 수 있다.
```

```python
for sido in data['광역시도'].unique():
    print(get_pop_max(sido),type(get_pop_max(sido)))
```

```python
# 새로운 DateFrame 객체 생성하기
max_pop_df = pd.DataFrame(columns=['광역시도','행정구역','인구수'])

for sido in data['광역시도'].unique():
    #print(get_pop_max(sido))
    max_pop_df = max_pop_df.append(get_pop_max(sido)) #리스트와 달리 append하고 끝나는게 아니라 대입 시켜줘야함
    
max_pop_df
```
