```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin

main_url = 'https://comic.naver.com/'
url = 'https://comic.naver.com/webtoon/weekday.nhn'
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
#print(soup)

webtoon_dict = {}
webtoon_list = []
for script in soup.select("#content .list_area"):
    #print(script)
    for i in script.select("a[href*=.nhn]"):
        link = urljoin(main_url, i['href'])
        title = i.text.strip()
        #print(link)
        #print(title)
        webtoon_dict['title'] = title
        webtoon_dict['link'] = link
        webtoon_list.append(webtoon_dict)
        webtoon_dict = {}

#print(webtoon_list)

for link in webtoon_list:
    a_link = link['link']
    print(a_link)
        
        
        
        
        
        
        
```
