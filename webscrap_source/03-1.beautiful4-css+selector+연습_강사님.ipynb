```python
from bs4 import BeautifulSoup 
#fp = open("books.html", encoding="utf-8")
soup = BeautifulSoup(book_html, "html.parser")
#<class 'bs4.element.Tag'> <class 'list'>
print(type(soup.select_one('#ge')),type(soup.select('#ge')))

#select_one, select
print(soup.select_one('#ge').string)
print(soup.select('#ge')[0].text)
#find, find_all
print(soup.find('li',attrs={"id" : "ge"}).text)
print(soup.find('li',attrs={"id" : "le","class":"second"}).text)

print(soup.find_all('li',attrs={"class" : "first"})[0].text)
for tag in soup.find_all('li',attrs={"class" : "first"}):
    print(tag.text)

# CSS 선택자로 검색하는 방법
print('===> CSS 선택자로 검색하는 방법')
sel = lambda q : print(soup.select_one(q).string)
sel("#ge")  
sel("li#ex")
sel("ul > li#le")
sel("#bible #nu")
sel("#bible > #de")
sel("ul#bible > li#ge")
sel("li[id='ex']")
sel("li:nth-of-type(5)")

# 그 밖의 방법 select 와 find_all은 list 를 반환한다.
print(soup.select(".first"))
```

```python
html="""
<html>
<body>
<div id="main-goods" role="page">
  <h1>과일과 야채</h1>
  <ul id="fr-list">
    <li class="red green" data-lo="ko">사과</li>
    <li class="purple" data-lo="us">포도</li>
    <li class="yellow" data-lo="us">레몬</li>
    <li class="yellow" data-lo="ko">오렌지</li>
  </ul>
  <ul id="ve-list">
    <li class="white green" data-lo="ko">무</li>
    <li class="red green" data-lo="us">파프리카</li>
    <li class="black" data-lo="ko">가지</li>
    <li class="black" data-lo="us">아보카도</li>
    <li class="white" data-lo="cn">연근</li>
  </ul>
</div>
<body>
</html>
"""
```

```python
from bs4 import BeautifulSoup 
soup = BeautifulSoup(html, "html.parser")

# 야채 만
print("--- 1.야채")
for ve_tag in soup.select("#ve-list li"):
    print(ve_tag.text)

# 과일 만
print("--- 2.과일")
for fr_tag in soup.select("#fr-list li"):
    print(fr_tag.text)    
    
# 야채 중에서 'ko'
print("--- 3. 야채 중에서 'ko'")
for ko_tag in soup.select("#ve-list li[data-lo='ko']"):
    print(ko_tag.text)

# 과일 중에서 'us'
print("--- 4. 과일 중에서 'us'")
for us_tag in soup.select("#fr-list li[data-lo='us']"):
    print(us_tag.text)
    
        
# 야채 중에서 'black'
print("--- 5. 야채 중에서 'black'")
for black_tag in soup.select('#ve-list li.black'):
    print(black_tag.text)
    
# 전체에서  red green 
print("--- 6. 전체에서  red green")
for red_tag in soup.select('li.red.green'):
    print(red_tag.text)

# 8번째 li
print("--- 7. 전체에서  8번째 li")
print(soup.select_one("li:nth-of-type(8)").string)

# 전체에서 'us'
print("--- 8. 전체에서 'us'")
for us_tag in soup.select("li[data-lo='us']"):
    print(us_tag.text)

# CSS 선택자로 추출하기
# print(soup.select("#ve-list > li[data-lo='us']")[1].string)
# print(soup.select("#ve-list > li.black")[1].string)
# print(soup.select_one("li:nth-of-type(8)").string)
# print(soup.select_one("#ve-list > li:nth-of-type(4)").string)
# # find 메서드로 추출하기 
# cond = {"data-lo":"us", "class":"black"}
# print(soup.find("li", cond).string)
# # find 메서드를 연속적으로 사용하기 
# print(soup.find(id="ve-list").find("li", cond).string)
```
